## Обучение

**Создание признаков**

1. разложим datetime на компоненты:
    - месяц
    - день
    - день недели
2. Отстающие значения, глубина задается параметром max_lag
3. Скользящее среднее, глубина задается параметром rolling_mean_size

```python
def make_features(data, max_lag, rolling_mean_size):
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

**LGBMRegressor**

Подберем параметры, которые минимизируют значения RMSE на тестовой выборке для LGBMRegressor

```python
seed=2

data_for_LGBM = data.copy()

def objective(params):
    
    est=int(params['n_estimators'])
    md=int(params['max_depth'])
    ml=int(params['max_lag'])
    rms=int(params['rolling_mean_size'])
    
    make_features(data_for_LGBM, ml, rms)
            
    train, test = train_test_split(data_for_LGBM, shuffle=False, test_size=0.1)
    train = train.fillna(value=0)
            
    train_features_LGBM = train.drop(['num_orders'], axis=1)
    train_target_LGBM = train['num_orders']
            
    test_features_LGBM = test.drop(['num_orders'], axis=1)
    test_target_LGBM = test['num_orders']
            
    model=LGBMRegressor(n_estimators=est,max_depth=md)
    model.fit(train_features_LGBM, train_target_LGBM)
    pred=model.predict(test_features_LGBM)
    score=mean_squared_error(test_target_LGBM, pred)**0.5
    return score

def optimize(trial):
    params={'n_estimators':hp.uniform('n_estimators', 10, 100),
           'max_depth':hp.uniform('max_depth',5,20),
           'max_lag':hp.uniform('max_lag',1,100),
           'rolling_mean_size':hp.uniform('rolling_mean_size',1,100)}
    
    best=fmin(fn=objective,space=params,algo=tpe.suggest,trials=trial,max_evals=50,rstate=np.random.default_rng(seed))
    
    return best

trial=Trials()
best=optimize(trial)
```

По итогам подбора получены параметры, которые минимизируют значения RMSE на тестовой выборке для LGBMRegressor

```python
best
```

Подготовим выборки согласно данных параметров

```python
make_features(data_for_LGBM, 57, 31)

train_for_LGBM, test_for_LGBM = train_test_split(data_for_LGBM, shuffle=False, test_size=0.1)
train_for_LGBM = train_for_LGBM.dropna()
            
train_features_LGBM = train_for_LGBM.drop(['num_orders'], axis=1)
train_target_LGBM = train_for_LGBM['num_orders']
            
test_features_LGBM = test_for_LGBM.drop(['num_orders'], axis=1)
test_target_LGBM = test_for_LGBM['num_orders']
```

**LinearRegression**

Подберем параметры, которые минимизируют значения RMSE на тестовой выборке для LinearRegression

```python
seed=2

data_for_LR = data.copy()

def objective(params):

    ml=int(params['max_lag'])
    rms=int(params['rolling_mean_size'])
    
    make_features(data_for_LR, ml, rms)
            
    train, test = train_test_split(data_for_LR, shuffle=False, test_size=0.1)
    train = train.fillna(value=0)
            
    train_features_LR = train.drop(['num_orders'], axis=1)
    train_target_LR = train['num_orders']
            
    test_features_LR = test.drop(['num_orders'], axis=1)
    test_target_LR = test['num_orders']
            
    model=LinearRegression()
    model.fit(train_features_LR, train_target_LR)
    pred=model.predict(test_features_LR)
    score=mean_squared_error(test_target_LR, pred)**0.5
    return score

def optimize(trial):
    params={'max_lag':hp.uniform('max_lag',1,100),
           'rolling_mean_size':hp.uniform('rolling_mean_size',1,100)}
    
    best=fmin(fn=objective,space=params,algo=tpe.suggest,trials=trial,max_evals=50,rstate=np.random.default_rng(seed))
    
    return best

trial=Trials()
best=optimize(trial)
```

По итогам подбора получены параметры, которые минимизируют значения RMSE на тестовой выборке для LinearRegression

```python
best
```

Подготовим выборки согласно данных параметров

```python
make_features(data_for_LR, 98, 50)

train_for_LR, test_for_LR = train_test_split(data_for_LR, shuffle=False, test_size=0.1)
train_for_LR = train_for_LR.dropna()
            
train_features_LR = train_for_LR.drop(['num_orders'], axis=1)
train_target_LR = train_for_LR['num_orders']
            
test_features_LR = test_for_LR.drop(['num_orders'], axis=1)
test_target_LR = test_for_LR['num_orders']
```

## Тестирование

**LGBMRegressor**

Обучим модель LGBMRegressor для полученных параметров и замерим время обучения

```python
%%time
model_LGBM = LGBMRegressor(n_estimators=84,max_depth=8)
model_LGBM.fit(train_features_LGBM, train_target_LGBM)
```

Получим предсказания на тестовой выборке и замерим время

```python
%%time
lgbm_prediction = model_LGBM.predict(test_features_LGBM)
```

```python
result = mean_squared_error(test_target_LGBM, lgbm_prediction)**0.5
print('RMSE на тестовой выборке для LGBMRegressor:', result.round(2))
```

**LinearRegression**

Обучим модель LinearRegression и замерим время обучения

```python
%%time
LR_model = LinearRegression()
LR_model.fit(train_features_LR, train_target_LR)
```

Получим предсказания на тестовой выборке и замерим время

```python
%%time
LR_prediction = LR_model.predict(test_features_LR)
```

```python
result = mean_squared_error(test_target_LR, LR_prediction)**0.5
print('RMSE на тестовой выборке для LinearRegression:', result.round(2))
```

**Лушчий показатель качества у модели LGBMRegressor при следующих параметрах:**

- max_lag: 57
- rolling_mean_size: 31

- max_depth: 8
- n_estimators: 84

- RMSE: 38.82

```python
plt.figure(figsize=(15, 10))
plt.plot(test_target_LGBM.index, test_target_LGBM, label='test_target_LGBM')
plt.plot(test_target_LGBM.index, lgbm_prediction, label='lgbm_prediction')
plt.plot(test_target_LR.index, LR_prediction, label='LR_prediction')
plt.xlabel('datetime')
plt.ylabel('num_orders')
plt.legend();
```

**Обе модели в основном спотыкаются при предсказании существенных всплесков заказов.**
